#Téléchargement de la base de données

Calepin pour la téléchargement et la décompression de la banque de données FMA dans Google Drive. Le service payant [Google One](https://one.google.com/) est requit pour avoir assez de stockage dans Google Drive.

##Setup

Ajouter Google Drive comme système de fichier pour Colab

```
from google.colab import drive
drive.mount('/content/gdrive')
```

##Téléchargement

Téléchargement du medatada de FMA

```
!wget -O gdrive/'My Drive'/"Projet d'intégration BdeB"/Téléchargement/fma_metadata.zip https://os.unil.cloud.switch.ch/fma/fma_metadata.zip
```

Téléchargement de la base de données fma_small

```
!wget -O gdrive/'My Drive'/"Projet d'intégration BdeB"/Téléchargement/fma_small.zip https://os.unil.cloud.switch.ch/fma/fma_small.zip
```

##Décompression

Décompression des téléchargements

```
!unzip gdrive/'My Drive'/"Projet d'intégration BdeB"/Téléchargement/fma_metadata.zip -d gdrive/'My Drive'/"Projet d'intégration BdeB"/FMA/fma_metadata
```

```
!unzip gdrive/'My Drive'/"Projet d'intégration BdeB"/Téléchargement/fma_small.zip -d gdrive/'My Drive'/"Projet d'intégration BdeB"/FMA/fma_small
```
